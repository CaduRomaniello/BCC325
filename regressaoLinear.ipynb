```python
!pip install word2number
import pandas as pd
import numpy as np
from sklearn import linear_model
import math
from word2number import w2n
```

```python
path_to_ds = "/content/drive/MyDrive/IA/candidatos.csv"
```

```python
df = pd.read_csv(path_to_ds)
```

```python
df
```

```python
i = 2
for i in range(2, len(df["experience"])):
  df["experience"][i] = w2n.word_to_num(df["experience"][i])
df
```

```python
median_experience = math.floor(df["experience"].median())
median_test_score = math.floor(df["test_score"].median())
```

```python
df["experience"] = df["experience"].fillna(median_experience)
df["test_score"] = df["test_score"].fillna(median_test_score)
df
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[["experience", "test_score", "interview_score"]], df["salary"])
```

```python
reg.coef_
```

```python
candidato = pd.DataFrame([[8, 7, 9]], columns=["experience", "test_score", "interview_score"])
```

```python
reg.predict(candidato)
```
